```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('data/Mall_Customers.csv')
df.head()
```

```python
# menyeleksi Fitur

X = df.iloc[:, 3:]
X.head()
```

```python
# Plot Data
# dikarenakan data 3 dimensi, 
# dengan plot cluster berdasarkan Annual Income dan Spending Score  saja

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], s = 100)
```

```python
# membuat Model KMeans
# mencoba menggunakan k=2 

from sklearn.cluster import KMeans
```

```python
# Inisiasi obyek KMeans

cl_kmeans = KMeans(n_clusters=2)
```

```python
# Fit dan predict model

y_kmeans = cl_kmeans.fit_predict(X)
```

```python
# Plot hasi cluster berdasarkan Annual Income dan Spending Score

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], s = 100, c=y_kmeans)
```

```python
# Plot centroid

centers = cl_kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
```

```python
# Cek Nilai SSE

print(f'Nilai SSE : {cl_kmeans.inertia_}')
```

```python
# Implementasi Metode Elbow


# List nilai SSE
sse = []

# Cari k terbaik dari 1-10
K = range(1,10)

# Cek nilai SSE setiap k

for k in K:
 kmeanModel = KMeans(n_clusters=k)
 kmeanModel.fit(X)
 sse.append(kmeanModel.inertia_)
```

```python
# Plotting the distortions

plt.figure(figsize=(8,4))
plt.plot(K, sse, "bx-")
plt.xlabel("k")
plt.ylabel("SSE")
plt.title("Metode Elbow untuk Mengetahui Jumlah k Terbaik")
plt.show()
```

```python
# Cek Nilai SSE setiap k

for idx, sse_val in enumerate(sse, start=1):
    print(f'k={idx}; SSE={sse_val}')
```

```python
# membuat Model KMeans
# mencoba menggunakan k=5 

from sklearn.cluster import KMeans
```

```python
# Inisiasi obyek KMeans

cl_kmeans = KMeans(n_clusters=5)
```

```python
# Fit dan predict model

y_kmeans = cl_kmeans.fit_predict(X)
```

```python
# Plot hasi cluster berdasarkan Sepal Length dan Sepal Width

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], s = 100, c=y_kmeans)
```

```python
# Cek Nilai SSE

print(f'Nilai SSE : {cl_kmeans.inertia_}')
```
